```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/europe.csv", delimiter=',', encoding='utf-8')
df
```

```python
to_normalize = ["Area", "GDP", "Inflation", "Life.expect", "Military", "Pop.growth", "Unemployment"]
for i in range(len(to_normalize)):
    df[to_normalize[i]] = (df[to_normalize[i]] - df[to_normalize[i]].mean()) / df[to_normalize[i]].std()
df
```

```python
from utils.kohonen.kohonen import build_kohonen_net
from utils.kohonen.neuron_neighbours import prop2iters_neighbour_radius
from utils.kohonen.neuron_lr import inv2iter_learning_rate
```

```python
kohonen_net = build_kohonen_net(df[["Area", "GDP", "Inflation", "Life.expect", "Military", "Pop.growth", "Unemployment"]], 2, 7*500, 
                                "sample no repos", "stochastic", 
                                prop2iters_neighbour_radius, inv2iter_learning_rate, 
                                "rectangular")
```

```python
austria = df[["Area", "GDP", "Inflation", "Life.expect", "Military", "Pop.growth", "Unemployment"]].iloc[0].to_numpy()
switzerland = df[["Area", "GDP", "Inflation", "Life.expect", "Military", "Pop.growth", "Unemployment"]].iloc[25].to_numpy()
luxemburg = df[["Area", "GDP", "Inflation", "Life.expect", "Military", "Pop.growth", "Unemployment"]].iloc[16].to_numpy()

greece = df[["Area", "GDP", "Inflation", "Life.expect", "Military", "Pop.growth", "Unemployment"]].iloc[9].to_numpy()
hungary = df[["Area", "GDP", "Inflation", "Life.expect", "Military", "Pop.growth", "Unemployment"]].iloc[10].to_numpy()
```
