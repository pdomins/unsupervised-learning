```python
df = pd.read_csv("../../data/europe.csv", delimiter=',', encoding='utf-8')
df
```

```python
df = df.set_index("Country")
df
```

```python
to_normalize = ["Area", "GDP", "Inflation", "Life.expect", "Military", "Pop.growth", "Unemployment"]
for i in range(len(to_normalize)):
    df[to_normalize[i]] = (df[to_normalize[i]] - df[to_normalize[i]].mean()) / df[to_normalize[i]].std()
df
```
